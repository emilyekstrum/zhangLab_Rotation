```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## -------------------- SVGS ---------------------

```python
# load in mouse dca outputs
svg_U = np.load("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/DCA/mouse_svg_U_DCA.npy") # gene loadings
svg_Z = np.load("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/DCA/mouse_svg_Z_DCA.npy") # cell embeddings
svg_dcovs = np.load("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/DCA/mouse_svg_dcovs_DCA.npy") # distance covariances per DCA dim
```

```python
# plost distance covariances
plt.figure(figsize=(8,6))
sns.barplot(x=np.arange(1, len(svg_dcovs)+1), y=svg_dcovs, color='skyblue')
plt.xlabel('DCA Dimension', fontsize=14)
plt.ylabel('Distance Covariance', fontsize=14)
plt.title('Distance Covariances for Mouse Brain SVG DCA Representation', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# plot cell embeddings in 2D
plt.figure(figsize=(8,6))
sns.scatterplot(x=svg_Z[:,0], y=svg_Z[:,1], s=10, color='navy', alpha=0.6)
plt.xlabel('DCA Dimension 1', fontsize=14)
plt.ylabel('DCA Dimension 2', fontsize=14)
plt.title('2D Scatter Plot of Mouse Brain SVG DCA Cell Embeddings', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()  
```

```python
# plot gene loadings for first two DCA dimensions
plt.figure(figsize=(8,6))
sns.scatterplot(x=svg_U[:,0], y=svg_U[:,1], s=10, color='darkgreen', alpha=0.6)
plt.xlabel('DCA Dimension 1 Loadings', fontsize=14)
plt.ylabel('DCA Dimension 2 Loadings', fontsize=14)
plt.title('Scatter Plot of Mouse Brain SVG DCA Gene Loadings', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

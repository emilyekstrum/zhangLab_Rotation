```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## -------------------- SVGS ---------------------

```python
# load in human DCA results
svg_U = np.load("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/DCA/human_svg_U_DCA.npy")
svg_Z = np.load("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/DCA/human_svg_Z_DCA.npy")
svg_dcovs = np.load("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/DCA/human_svg_dcovs_DCA.npy")
```

```python
# plot distance covariances
plt.figure(figsize=(6,4))
sns.barplot(x=np.arange(1, svg_dcovs.shape[0]+1), y=svg_dcovs, color='skyblue')
plt.xlabel("DCA Dimension")
plt.ylabel("Distance Covariance")
plt.title("Distance Covariances per DCA Dimension (Human Brain)", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# plot cell embeddings in 2D
plt.figure(figsize=(8,6))
sns.scatterplot(x=svg_Z[:,0], y=svg_Z[:,1], s=10, color='navy', alpha=0.6)
plt.xlabel('DCA Dimension 1', fontsize=14)
plt.ylabel('DCA Dimension 2', fontsize=14)
plt.title('2D Scatter Plot of Human Brain SVG DCA Cell Embeddings', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# plot gene loadings for first two DCA dimensions
plt.figure(figsize=(8,6))
sns.scatterplot(x=svg_U[:,0], y=svg_U[:,1], s=10, color='darkgreen', alpha=0.6)
plt.xlabel('DCA Dimension 1 Loadings', fontsize=14)
plt.ylabel('DCA Dimension 2 Loadings', fontsize=14)
plt.title('Scatter Plot of Human Brain SVG DCA Gene Loadings', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

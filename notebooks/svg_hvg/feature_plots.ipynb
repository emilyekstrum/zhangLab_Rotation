# Expression feature plots 

```python
import scanpy as sc
import matplotlib.pyplot as plt
```

```python
# load in SVGs from csv files
mouse_svg_df = sc.read_csv("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/svgs/nnSVG_mouse_svg_gene_cell_matrix.csv", first_column_names=True)
human_svg_df = sc.read_csv("/Users/emilyekstrum/repos/zhangLab_Rotation/data/processed/svgs/nnSVG_human_svg_gene_cell_matrix.csv", first_column_names=True)
```

```python
# convert to pandas DataFrame for easier manipulation
mouse_svg_df = mouse_svg_df.to_df() # gene x spot
human_svg_df = human_svg_df.to_df() # gene x spot
```

```python
# expression histograms for top 20 SVGs in mouse
top20_svg_mouse_genes = mouse_svg_df.index[:20]
mouse_svg_df.loc[top20_svg_mouse_genes].T.hist(bins=30, figsize=(15,10))
plt.suptitle("Expression Histograms for Top 20 Mouse SVGs", fontsize=16)
# label each x and y axis
for ax in plt.gcf().axes:
    ax.set_xlabel("Expression Level")
    ax.set_ylabel("Number of Spots")
plt.tight_layout()
plt.show()
```

```python
# expression histograms for top 20 SVGs in human
top20_svg_human_genes = human_svg_df.index[:20]
human_svg_df.loc[top20_svg_human_genes].T.hist(bins=30, figsize=(15,10))
plt.suptitle("Expression Histograms for Top 20 Human SVGs", fontsize=16)
# label each x and y axis
for ax in plt.gcf().axes:
    ax.set_xlabel("Expression Level")
    ax.set_ylabel("Number of Spots")
plt.tight_layout()
plt.show()
```
